```python
df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','', regex = True)
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"]
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df["Phone_Number"]
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '')
df
```

```python
df
```

```python
df["Address"].str.split(',', n = 2, expand = True)
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n = 2, expand = True)
df
```

```python
df = df.drop_duplicates()
df
```

```python
df.loc[:, "Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y')
df.loc[:, "Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
df
```

```python
df.loc[:, "Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
df.loc[:, "Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
df
```

```python
#df = df.replace('N/a', '')
#df = df.replace('N/a', '')

df = df.fillna('')
df
```

```python
# removes rows with Y in Do_Not_Contact
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)

df
```

```python
# removes rows without phone numbers
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

df
```

```python
df = df.reset_index(drop = True)
df
```
